```python
#importing packages.
import pandas as pd
from collections import Counter
```

```python
#DATA COLLECTION - fetching raw database in csv format from url.
url = 'https://raw.githubusercontent.com/coinmetrics-io/data/master/csv/btc.csv'
raw_data = pd.read_csv(url)
raw_data.tail()
```

```python
#Info from raw_data
#43 columns and 4329 rows
#All columns are floats, except the first column 'time' (object type), which is the date.
#Column PriceUSD, the price of bitcoin in USD, has 3768 entries.
#For modeling purposes we may have to drop the rows where PriceUSD is null.
raw_data.info()
```

```python
#The variable definitions for each column is in the 'metrics_info.csv' file.
metrics_info = pd.read_csv('../data/raw/metrics_info.csv').transpose()
#Here I also name the columns with the id row and "erase" the 'id' row.
metrics_info.columns = metrics_info.loc['id']
metrics_info = metrics_info.loc['name': 'description']
metrics_info
```

```python
#null values of PriceUSD column
Counter(pd.isnull(raw_data['PriceUSD']))
```

```python
#Sumamry statistics
raw_data.describe()
```

```python
#setting date column as index and droping it from columns
raw_data.index = pd.to_datetime(raw_data['time'])
raw_data = raw_data.drop('time', axis='columns')
raw_data.tail()
```

```python
#drop rows in 'PriceUSD' with NaN values
data = raw_data.dropna(subset=['PriceUSD'])
data.head()
```

```python
data.describe()
```

```python
#Most variables have now 100% non-null values.
data.info()
```

```python
#Most columns have now 100% non-null values.
#Columns with less than 100% non-null values are data whose calculation depends on trailing moving averages
#of 30-365 days. These values will be set to zero.
data_pct_non_null = data.count()/len(data)*100
data_pct_non_null
```

```python
#replacing non-null values with zeros.
data = data.fillna(0)
data.head()
data_pct_non_null = data.count()/len(data)*100
data_pct_non_null
```

```python
#Additional dataframes for monthly and annual data will be created to analyse in next stages
#if there are clearer paterns.
#agg_dict shows the agg fnction for each column in data.

agg_dict = {'AdrActCnt': 'mean', 'BlkCnt': 'sum', 'BlkSizeByte': 'sum', 'BlkSizeMeanByte': 'mean', 'CapMVRVCur': 'mean',
       'CapMrktCurUSD': 'mean', 'CapRealUSD': 'mean', 'DiffMean': 'mean', 'FeeMeanNtv': 'mean', 'FeeMeanUSD': 'mean',
       'FeeMedNtv': 'mean', 'FeeMedUSD': 'mean', 'FeeTotNtv': 'sum', 'FeeTotUSD': 'sum', 'HashRate': 'mean',
       'IssContNtv': 'sum', 'IssContPctAnn': 'mean', 'IssContUSD': 'sum', 'IssTotNtv': 'sum', 'IssTotUSD': 'sum',
       'NVTAdj': 'mean', 'NVTAdj90': 'mean', 'PriceBTC': 'mean', 'PriceUSD': 'mean', 'ROI1yr': 'mean', 'ROI30d': 'mean',
       'SplyCur': lambda x: x.iloc[-1], 'SplyExpFut10yrCMBI': lambda x: x.iloc[-1], 'SplyFF': lambda x: x.iloc[-1], 'TxCnt': 'sum',
       'TxTfrCnt': 'sum', 'TxTfrValAdjNtv': 'sum' , 'TxTfrValAdjUSD': 'sum', 'TxTfrValMeanNtv': 'mean',
       'TxTfrValMeanUSD': 'mean', 'TxTfrValMedNtv': 'mean', 'TxTfrValMedUSD': 'mean', 'TxTfrValNtv': 'sum',
       'TxTfrValUSD': 'sum', 'VtyDayRet180d': 'mean', 'VtyDayRet30d': 'mean', 'VtyDayRet60d': 'mean'}
    
#resample data by month.
monthly_data = data.resample('M').agg(agg_dict)
monthly_data.tail()
```

```python
#resample data by year.
annual_data = data.resample('A').agg(agg_dict)
annual_data
```

```python
annual_prices = annual_data['PriceUSD']
annual_prices
```

```python

#The goal here is to create a column that accounts for the halving periods of bitcoin block subsidy.

#First we create the column 'block_height', which is the cumulative sum of blocks produced each day.
data['block_height'] = raw_data['BlkCnt'].cumsum(axis=0)
print(data['block_height'])

#We create the 'halving_period' column by dividing the 'block_height' by 210,000 
data['halving_period'] = data['block_height'].apply(lambda x: x // 210000)
data['halving_period']

#The 'halving_period' column will allow us to analize the data taking into account the halving periods.
```
