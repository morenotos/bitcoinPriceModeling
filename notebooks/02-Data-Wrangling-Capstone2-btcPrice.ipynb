```python
#importing packages.
import pandas as pd
from collections import Counter
```

```python
#DATA COLLECTION - fetching raw database in csv format from url.
url = 'https://raw.githubusercontent.com/coinmetrics-io/data/master/csv/btc.csv'
raw_data = pd.read_csv(url)
raw_data.tail()
```
